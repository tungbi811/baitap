```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
def MSE(a,b,X,y):
    y_hat = a*X+b
    m = len(y)
    return np.sum((y-y_hat)**2)/m
```

```python
data = pd.read_csv('data/data_square.csv')

X = data["Diện tích"]
y = data["Giá nhà"]

X = np.array([X]).T
y = np.array([y]).T
```

```python
plt.figure("Bai 3")
plt.scatter(X,y)
plt.show()
```

```python
model = LinearRegression()
model.fit(X,y)
a = model.coef_
b = model.intercept_
print("Coef = ",a)
print("Intercept = ",b)
print("MSE = ",MSE(a,b,X,y))
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),'-r')
print("Model is not good because it is under fit!")
plt.show()
```
