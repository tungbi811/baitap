```python
import numpy as np
import matplotlib.pyplot as plt

dic = {'10-20':[10, 15], '20-30':[50, 75], '30-40':[40, 55], '40-50':[30, 25]} 
age = list(dic.keys())
bike = list(dic.values())
age = np.array(age)
bike = np.array(bike)

plt.figure(1)
plt.bar(age,bike[:,0],width=-0.4,color='red',align='edge')
plt.bar(age,bike[:,1],width=0.4,color='blue',align='edge')
plt.legend(["Male","Female"])
plt.xlabel("Age")
plt.ylabel("Number of people ride bikes")

plt.figure(2)
plt.bar(age,bike[:,1],width=0.4,color='blue')
plt.bar(age,bike[:,0],width=0.4,color='red')
plt.legend(["Female","Male"])
plt.xlabel("Age")
plt.ylabel("Number of people ride bikes")
plt.show()
```
