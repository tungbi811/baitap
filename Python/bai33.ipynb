```python
from numpy import dtype
import pandas as pd
import numpy as np
```

```python
records = pd.read_csv('data/housing.data',sep='\s+')
print("Number of rows:",records.shape[0])
print("Number of cols:",records.shape[1])
```

```python
housing_name = []
names = open('data/housing.names',"r",encoding='cp1252')
for line in names:
    if line.startswith("7. Attribute Information:"):
        while True:
            line = next(names)
            line = line.rstrip().strip()
            s = line.split(" ")
            if line.startswith("8. Missing Attribute Values:"):
                break
            if line != "":
                housing_name.append(s[1])

data = np.array(records)
df = pd.DataFrame(data=data,columns=housing_name)
print(housing_name)
print(df)
```
